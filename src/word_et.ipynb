```python
import sys
import pickle
import json
import re
import numpy as np
from scipy.sparse import csr_matrix
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
```

```python
def feature_transform(train_file, words):
    
    words_len = len(words)
    row_idx = []
    column_idx = []
    data = []
    y = []
    ctr = 0
    feature_length = words_len # Per head

    f = open(train_file)

    for line in f:
            if(line.rstrip()):
                line = re.sub("\s+"," ",line)
                line1 = line.split(";")

                a1 = line1[0].split(" ")
                a2 = line1[1].split(" ")
                a3 = line1[2].split(" ")
                a4 = line1[3].strip()


                if a3[1] != "U":
                    
                    ctr += 1
                    if(a1[0] == "H"):
                        column_idx.append(words.index(a1[1]))

                    elif(a1[0] == "ROOT"):
                        column_idx.append(words.index("ROOT"))

                    row_idx += [ctr-1]*2
                    data += [1] *2
                    column_idx.append(feature_length + words.index(a2[2]))

                    y.append(a4)

    f.close()

    X = csr_matrix((data, (row_idx, column_idx)), shape=(ctr,2*(words_len)))
    
    
    return X, y
```

```python
listfile = "data_tokens.json"
f = open(listfile)
data = json.load(f)
f.close()

words = data["words"]

train_file = 'training_data.txt'
test_file = "testing_data.txt"

X_train, y_train = feature_transform(train_file, words)
X_test, y_test = feature_transform(test_file, words)


model = LinearSVC()
model.fit(X_train, y_train)

pred_train = model.predict(X_train)
pred_test = model.predict(X_test)
```

```python
print(classification_report(y_train, pred_train))
```

```python
print(classification_report(y_test, pred_test))
```
